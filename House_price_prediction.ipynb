```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/house_price.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df['apartment'] = label.fit_transform(df['apartment'])
```

```python
df['parking']=label.fit_transform(df['parking'])
```

```python
df['location_mumbai']=label.fit_transform(df['location_mumbai'])
df.head()
```

```python
x = df.iloc[:,1:5]
```

```python
y = df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
pricemodel = RandomForestRegressor()
```

```python
pricemodel.fit(x_train, y_train)
```

```python
x_test
```

```python
y_pred=pricemodel.predict(x_test)
print(y_pred)
y_test
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print('testing score:',pricemodel.score(x_test, y_test)*100)
print('Actual op vs pred op:',mean_absolute_percentage_error(y_test, y_pred))
print('training score:',pricemodel.score(x_train, y_train)*100)
```
